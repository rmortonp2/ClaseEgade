```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Canada.xlsx',
             sheet_name = 'Canada by Citizenship',
             skiprows = range(20),
                   skipfooter = 2,
            )
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Type', 'Coverage',   'AREA', 'REG','DEV',  'DevName'],
        axis = 1,
        inplace = True )
df.head()
```

```python
df.columns
```

```python
df.rename(columns = {'OdName':'Country',
                    'AreaName':'Continent',
                     'Regname':'Region',
                    },
          inplace = True)
df.head()
```

```python
df.columns
```

```python
all(isinstance(i,str)
    for i in df.columns)
```

```python
df.columns = list (map(str,df.columns))
```

```python
all(isinstance(i,str)
    for i in df.columns)
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.head()
```

```python
df.set_index('Country',
             inplace = True)
df.head(2)
```

```python
df['Total'] = df.iloc[:,2:].sum(axis = 1)
df.head(2)
```

```python
years = list(map(str,range(1980,2014)
                )
            )
years
```

```python
pip install matplotlib
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.available
mpl.style.use('ggplot')
```

```python
italia = df.loc['Italy',years]
italia
```

```python
line_graph_italy = italia.plot(
                                figsize = (15,10)
                               )
plt.title('Inmigration from Italy during 1980 - 2013 to Canada',fontsize = 15)
plt.ylabel ('Number of inmigrants',fontsize = 15)
plt.text(7,1190,'Reason', fontsize = 15)
plt.xlabel ('Years', fontsize = 15)
```

```python
two = df.loc[{'India','China'},years]
two = two.transpose()
```

```python
line_graph_two = two.plot(
                                figsize = (15,10)
                               )
```

# Seleccionar los mayores 5

```python
top = df.sort_values(by = 'Total',
                     ascending = False,
                     axis = 0,
                    )
top
```

```python
top6 = top.head(6)
top6
```

```python
top6 = top6[years].transpose()
top6
```

```python
line_graph_top = top6.plot(figsize = (15,10))
plt.title('A title',fontsize = 15)
plt.ylabel ('# Inmigrants',fontsize = 15)
plt.xlabel ('Years', fontsize = 15)
```

# Area plots (staked line plot)

```python
top6.plot(kind = 'area',
          figsize = (15,8),
          stacked = False, # do not aggregating the values
          alpha = 0.15 # intensity of the area in graph
         )
plt.title('Title',fontsize = 18)
```

# Histogram

```python
df['2012'].plot(kind = 'hist')
```

```python
np.histogram(df['2012'])
```

```python
count,bin_edges = np.histogram(df['2012'])
print('counts \t:',count)
print('Bin edges \t:',bin_edges)
```

```python
xmin = bin_edges[0] # - 1000
xmax = bin_edges[-1] # +1000
print(xmin,xmax)
```

```python
df['2012'].plot( kind = 'hist',
                figsize = (15,8),
                xticks = bin_edges,
               xlim = (xmin,xmax)
               )
```

```python
fav_countries = ['Denmark','Norway','Sweden']
fav = df.loc[fav_countries,['2012','2013']].transpose()
```
